## Abundances Throughout Thermal Pulse

From the Kippenhahn and luminosity plots:  
* Start of the AGB before any pulses: model 1700
* I'm not sure about the pulse from 2100-2300 
   * (shows up in L but doesn't produce convective pocket). 
   * Ignore this one and work with the first convective pocket. 
* Right before a definite pulse: 2500
* Peak of that pulse: 2660
* Right after that pulse: 2750
* Interpulse (LHe valley): 2930  

I will plot the pre-pulse AGB zoomed out to show overview (location of H/He-free core, H/He-burning shells, etc)  
Everything else will be zoomed in on the pulse region. 

```python
species=['H-1','He-4','C-12','C-13','N-14','O-16']
```

```python
ifig=104;close(ifig);figure(ifig)
star.abu_profile(isos=species, ifig=ifig, fname=1700, logy=True)
title('Abundances on AGB Before Pulses Begin (c=1700)')
xlim(-0.1,2.5)
ylim(-100,5)
```

All of the abundance plots will have this overall shape,  
so the rest will be plotted just to show the interesting region around the H and He shells. 

```python
ifig=105;close(ifig);figure(ifig)
star.abu_profile(isos=species, ifig=ifig, fname=2500, logy=True)
title('Abundances Preceding an Early Pulse (c=2500)')
xlim(1.06, 1.07)
ylim(-30,5)
```
