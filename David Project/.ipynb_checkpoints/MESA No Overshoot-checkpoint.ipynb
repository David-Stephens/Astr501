```python
%pylab 
from nugridpy import nugridse as mp
from nugridpy import mesa as ms
import matplotlib.pyplot as plt
import numpy as np

data_dir="/user/ast501/dstephens/1M_pre_ms_to_wd/LOGS/"    

ms.set_nugrid_path(data_dir)
mp.set_nugrid_path(data_dir)
```

```python
star = ms.star_log(data_dir)
```

```python
# create a kippenhahn diagram
plt.figure(1)
star.kippenhahn_CO(1,'model')

# look at the convection zones
plt.xlim([14000,16000])
plt.ylim([0.5,0.58])
```

```python
# grab the mesa profile data (15250)
star_prof = ms.mesa_profile(data_dir,num=15250)
```

```python
# check the dels for convection as well as any other sources of D
mass = star_prof.get('mass')
diffusion_mlt = star_prof.get('log_D_conv')
diffusion_mix = star_prof.get('log_D_thrm')
del_rad = star_prof.get('gradr')
del_ad = star_prof.get('grada')

plt.figure(3)
plt.plot(mass,diffusion_mlt,'b-')
plt.plot(mass,diffusion_mix,'k-')

# create a new axis for the dels
ax1 = plt.gca()
ax2 = ax1.twinx()

ax2.plot(mass,del_rad,'r--')
ax2.plot(mass,del_ad,'g--')

# limit to the first convection zone
ax1.set_xlim([0.52,0.6])
ax1.set_ylim([0,20])

# limit the dels
ax2.set_ylim([0,10])
```

```python
# check the entropy
entropy = star_prof.get('entropy')

plt.figure(2)
plt.plot(mass,diffusion_mlt)

# create a new axis for plotting entropy
ax1 = plt.gca()
ax2 = ax1.twinx()

ax2.plot(mass,entropy,'r--')

# limit to the first convection zone
ax1.set_xlim([0.52,0.6])
ax1.set_ylim([0,20])

# labels
ax1.set_ylabel('Diffusion Coefficient')
ax2.set_ylabel('Entropy')

# get axis xlim
conv_xlim = ax1.get_xlim()
```

```python
# different model number, not one with the convection zone but c13 pocket
star_prof2 = ms.mesa_profile(data_dir,15000)
```

```python
# plot some of the abundances
o16 = star_prof2.get('o16')
h1 = star_prof2.get('h1')
he4 = star_prof2.get('he4')
n14 = star_prof2.get('n14')
ne22 = star_prof2.get('ne22')
mass = star_prof2.get('mass')

plt.figure(4)
plt.plot(mass,np.log10(h1),label='h1')
plt.plot(mass,np.log10(he4),label='he4')
plt.plot(mass,np.log10(o16),label='o16')
plt.plot(mass,np.log10(ne22),label='ne22')
plt.plot(mass,np.log10(n14),label='n14')

# set limits
plt.ylim([-7,0])
plt.xlim(conv_xlim)

plt.legend()
```
