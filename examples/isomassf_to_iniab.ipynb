```python
%pylab
from nugridpy import ppn 
from nugridpy import ascii_table as at
```

```python
rhe=ppn.abu_vector('/user/ast501/digbyr/NuPPN/frames/ppn/RUN_Heburn')
```

```python
cycle=30
names=rhe.get('ISOTP',fname=cycle)
```

```python
abu=rhe.get('ABUNDANCE_MF',fname=cycle).tolist()
```

```python
z=array(rhe.get('Z',fname=cycle),dtype=int).tolist()
```

```python
data=[z,names,abu]
```

```python
dcols=['Z', 'species','mass fraction']
hd=["info about this initial abundance"]
outfile='iniab.dat'
%rm iniab.dat
at.write(outfile,hd,dcols,data)
```
