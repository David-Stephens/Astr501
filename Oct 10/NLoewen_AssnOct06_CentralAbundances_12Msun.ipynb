## Derive the temperature radiative gradient for the stellar interior. Clearly state your assumptions
We first consider the mean free path of a photon $l_{ph}$ for a stellar interior:
$l_{ph} = \frac{1}{\kappa \rho} \approx 1cm$


where $\kappa$ is opacity and $\rho$ is density.  Since this is a very small mean free path compared to eg. pressure scale heights in the star, we assume we can use a diffusion approximation for energy flux in the star.  We also as a result consider the interior to be optically thick, which allows us to use a maxwell-boltzmann energy distribution to describe the state of photons in the star.

The energy flux in the star is:

$j = -D\nabla n$

,where j is flux, D is the diffusion coefficient, $\nabla$ is internal energy gradient, and n is the photon density.


The diffusion coefficient is defined as $D = \frac{l v}{3}$ where $l$ is mean free path of the medium particle and $v$ is their velocities.  Using our above result for $l_{ph}$ and the fact that for photons $v=c$ this becomes:

$D = \frac{l v}{3} = \frac{l_{ph} v}{3} = \frac{c}{3\rho\kappa}$


The internal energy gradient $\nabla$ is determined from the maxwell-boltzmann energy distribution $u = aT^4$ where a is the radiative energy constant(?), which gives us:

$\nabla = \frac{\delta}{\delta r}aT^4 = 4aT^3\frac{\delta T}{\delta r}$


The flux $j$ can be written in terms of total luminosity of the star $L$ as:

$j = \frac{L}{4\pi r^2}$


Putting all these together gives us the equation:

$\frac{L}{4\pi r^2} = -\frac{c}{3\rho\kappa}4aT^3\frac{\delta T}{\delta r}$

where solving for $\frac{\delta T}{\delta r}$ gives us the radiative temperature gradient:

$\frac{\delta T}{\delta r} = -\frac{3}{16\pi a c}\frac{\kappa\rho L}{r^2 T^3}$
