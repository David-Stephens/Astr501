## Mesa Profiles

Get the mesa profile for the 3 M$_{\odot}$ star and look at the $\nabla$'s

```python
# model number to look at
model_number = 69000
data = ms.mesa_profile(mass=3,Z=0.02,num=model_number)

# check the dels
dela = data.get('grada')
delr = data.get('gradr')
delt = data.get('gradT')

# grab the radius
r = data.get('logR')

# grab the mass
mass = data.get('mass')

# get the entropy
entropy = data.get('entropy')
```

```python
# create plot
plt.figure(1)

# plot
plt.plot(mass,dela)
plt.plot(mass,delr)
plt.plot(mass,delt,':')

# create ylim and xlim
plt.ylim([0,1])
plt.xlim([0,1.75])
```

```python
# create an entropy plot
plt.figure(2)

# plot
plt.plot(mass,entropy)

# create xlim
plt.xlim([0,1.75])
```
